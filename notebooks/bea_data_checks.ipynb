```python
# Imports
import pandas as pd
```

```python
# Load processed data
df = pd.read_csv('../../insights-project-v2/data/processed/processed_bea.csv')
print(f"Loaded {len(df)} rows and {df.shape[1]} columns from processed BEA data")
df.head(5)
```

```python
# Check column names
print("Columns:", df.columns.tolist())

# Check data types
print(df.dtypes)

# Missing values per column
missing = df.isna().sum()
print("Missing values per column:\n", missing)

# Missing values in year columns
year_cols = [col for col in df.columns if col.isdigit()]
print("Missing values in year columns:\n", df[year_cols].isna().sum())

# Check year coverage
print("Years in dataset:", sorted([col for col in df.columns if col.isdigit()]))

# Unique GeoNames (should just be West Virginia or relevant regions)
print(df['geoname'].unique())

# Sample of industries
print(df['industryclassification'].unique()[:10])

# Check total GDP by year
total_gdp = df[year_cols].sum()
print(total_gdp)
```
